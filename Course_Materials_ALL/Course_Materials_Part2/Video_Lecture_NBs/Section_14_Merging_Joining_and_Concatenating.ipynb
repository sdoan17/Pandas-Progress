# Outer Join

```python
men2004 = pd.read_csv("men2004.csv")
```

```python
men2008 = pd.read_csv("men2008.csv")
```

```python
men2004.shape
```

```python
men2008.shape
```

```python
len(men2008) + len(men2004)
```

```python
men0408= men2004.merge(men2008, how = "outer", on = "Athlete", suffixes= ("_2004", "_2008"), indicator= True)
```

```python
men0408._merge.value_counts()
```

# Inner Join

```python
men2004.merge(men2008, how = "inner",on = "Athlete" ,suffixes= ("_2004", "_2008"), indicator= True)
```

```python
men0408_inner = men2004.merge(men2008, how = "inner",on = "Athlete" ,suffixes= ("_2004", "_2008"), indicator= True)
```

# Outer Join w/o intersection

```python
men0408.loc[men0408._merge != "both"]
```

# Left Join without Intersection

```python
men0408_left_wo_inter = men2004.merge(men2008, how = "left", on = "Athlete", suffixes= ("_2004", "_2008"), indicator= True)
```

```python
men0408_left_wo_inter
```

```python
men0408_left_wo_inter.loc[men0408_left_wo_inter._merge != "both"]
```

# Joining on different Column Lables & Indexes

```python
men2004 = pd.read_csv("men2004.csv")
men2008 = pd.read_csv("men2008.csv")
```

```python
men2004.columns = ["Name", "Medals"]
```

```python
men2008
```

```python
men0408=men2004.merge(men2008, how ="outer", left_on="Name",right_on="Athlete", suffixes= ["_2004", "_2008"], indicator= True)
```

men0408.fillna(men0408.Athlete, inplace= True)

```python
men0408
```

```python
men0408.drop(columns= ["Athlete","_merge"], inplace=True)
```

```python
men2004.head()
```

```python
men2008.set_index("Athlete", inplace= True)
```

```python
men2008.head()
```

```python
men2004.merge(men2008, how ="outer", left_on="Name",right_index=True, suffixes= ["_2004", "_2008"], indicator= True)
```
