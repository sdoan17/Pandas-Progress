# Concatenating DataFrames vertically

```python
import pandas as pd
```

```python
cars = pd.read_csv("cars_clean.csv")
```

```python
cars.head(10)
```

```python
cars.origin.value_counts()
```

```python
cars_jap = pd.read_csv("cars_jap.csv")
```

```python
cars_jap.head(10)
```

```python
cars_jap.insert(7, "origin", "japan")
```

```python
cars_jap
```

```python
cars_jap.columns = cars.columns
```

```python
cars_all = pd.concat([cars, cars_jap], ignore_index= True)
```

```python
cars_all
```

```python
cars_all = cars_all.sort_values("model_year").reset_index(drop = True)
```

# Left Join

```python
summer = pd.read_csv("summer.csv")
dic = pd.read_csv("dictionary.csv")
```

```python
dic
```

```python
summer
```

```python
summer_new = summer.merge(dic, how = "left", left_on= "Country",right_on="Code")
```

```python
summer_new
```

```python
summer_new.info()
```

```python
summer_new.loc[summer_new.Code.isnull()]
```
