# Customization

```python
titanic.age.plot(figsize = (12,8), fontsize = 13, c = "orange", 
                 xticks = xticks, yticks = yticks,rot = 45)
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
#plt.grid()
plt.show()
```

# Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.age.value_counts(bins = 10)
```

```python
titanic.age.plot(kind = "hist", figsize = (12,8), fontsize = 15, bins = 80, density = True)
plt.show()
```

```python
titanic.age.hist(figsize = (12,8), bins = 80, xlabelsize= 15, ylabelsize= 15)
plt.show()
```
