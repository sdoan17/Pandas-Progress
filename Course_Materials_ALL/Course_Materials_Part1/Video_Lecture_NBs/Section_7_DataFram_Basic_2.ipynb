
# Filtering DataFrams with many Conditions (AND)

```python
mask1 = titanic.sex == "male"
mask1
```

```python
mask2 = titanic.age >14
mask2.head()
```

# Advanced Filtering with between(), isin(), and ~(not)

```python
summer = pd.read_csv("summer.csv")
```

```python
og_1988 = summer.loc[summer.Year == 1988]
og_1988
```

```python
og_1988.info()
```

```python
og_since1922 = summer.loc[summer.Year >= 1992]
og_since1922
```

```python
summer.Year.between(1960,1969).head()
```

```python
og_60s = summer.loc[summer.Year.between(1960,1969,inclusive="both")]
og_60s
```

```python
my_fav_games = [1972,1996]
```

```python
summer.Year.isin(my_fav_games).head(20)
```

```python
og_72_96 = summer.loc[summer.Year.isin(my_fav_games)]
og_72_96
```

```python
og_not_72_96 = summer.loc[~summer.Year.isin(my_fav_games)]
og_not_72_96.head()
```

```python
og_not_72_96.Year.unique()
```
