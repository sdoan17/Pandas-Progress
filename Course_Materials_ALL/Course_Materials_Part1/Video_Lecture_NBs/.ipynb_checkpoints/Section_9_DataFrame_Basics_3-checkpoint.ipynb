```python
titanic.sort_values(by = "age", ascending= False, inplace= False)
```

```python
titanic
```

```python
titanic.sort_values(by = ["pclass", "sex", "age"], ascending= [True, True,False], inplace= True) 
#pclass and sex are in ascending, age is in decending. Still prioritize the first element
```

```python
titanic
```

```python
titanic.sort_index(ascending= True, inplace= True) #reset the dataframe by sort the index
```

```python
titanic
```

```python
titanic.sort_values(by = "age", ignore_index= True) 
# sort the dataframe and also reset the index accroding to dataframe
```

# Ranking DataFrames with rank()

```python
sales = pd.Series([15,32,45,21,55,15,0], index = ["Mo", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
```

```python
sales
```

```python
sales.sort_values(ascending=False)
```

```python
sales.rank().sort_values(ascending=False)
```

```python
sales.rank(ascending=False).sort_values(ascending=True)
```

```python
sales.rank(ascending=False, method = "min").sort_values(ascending=True)
```

```python
sales.rank(ascending=False, method = "min", pct = True).sort_values(ascending=True)
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.fare.rank(ascending= False)
```

```python
titanic["fare_rank"] = titanic.fare.rank(ascending=False, method = "min")
```

```python
titanic.head()
```

```python
titanic.sort_values(by = "fare", ascending= False)
```

# nunique(), nlargest() and nsmallest() with DataFrames

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.age.unique()
```

```python
titanic.nunique(axis = 0, dropna= False)
```

```python
titanic.nlargest(n = 5, columns = "fare") #n is the number of rows ordered by the columns
```

# Summary Statistics and Accumulations

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.describe()
```

```python
titanic.count(axis = "columns") # count the number of non-missing value per row
```

```python
titanic.sum(axis = 0, numeric_only=True)
```

```python
titanic.sum(axis = 0, numeric_only=True).loc["survived"]
```

```python
titanic.corr(numeric_only=True)
```

```python
titanic.survived.corr(titanic.pclass)
```
