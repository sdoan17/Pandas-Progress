# Counting Medals

```python
summer
```

```python
summer["Medal_Count"] = summer.groupby(["Year", "Sport", "Discipline", "Event", "Gender_Event"]).Medal.transform("count")
```

```python
summer.Medal_Count.value_counts().sort_index()
```

# TeamEvent and Single Event

```python
summer
```

```python
summer["Team_Event"] = pd.Series(np.where(summer.Medal_Count > 5, "Yes", "No"))
```

```python
summer.loc[summer.Team_Event.isna()] = "No"
```

Since there are team events, there are alot of duplicate row

```python
summer.drop(columns = "index", inplace= True)
```

```python
summer
```

```python
single = summer.loc[summer.Team_Event == "No"].copy()
single
```

```python
team = summer.loc[summer.Team_Event == "Yes"].copy()
team
```

```python
team.drop_duplicates(subset = ["Year", "Sport", "Discipline", "Country", "Event", "Gender_Event", "Medal"], inplace = True)
```

```python
summer_new = pd.concat([single,team])
```
