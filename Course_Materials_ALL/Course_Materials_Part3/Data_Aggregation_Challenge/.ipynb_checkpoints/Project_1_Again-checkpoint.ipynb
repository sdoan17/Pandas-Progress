# Determine Event_Gender

```python
summer
```

```python
summer["Gender_Event"] = summer.Gender
```

```python
mixEvent = summer.Event.str.lower().str.contains("mixed")
```

```python
pairsEvent = summer.Event.str.lower().str.contains("pairs")
```

```python
equesEvent = summer.Sport.str.lower().str.contains("equestrian")
```

```python
sailingEvent = ((summer.Sport.str.lower().str.contains("sailing")) & (summer.Year < 1988))
```

```python
summer.loc[mixEvent|pairsEvent|equesEvent|sailingEvent, "Gender_Event"] = "X"
```

```python
summer.loc[summer.Gender_Event == "X"]
```
