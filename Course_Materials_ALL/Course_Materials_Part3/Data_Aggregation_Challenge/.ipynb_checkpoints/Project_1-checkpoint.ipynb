# Seperating Event Gender

```python
summer
```

```python
#event contains the string "mixed"
summer.Event.str.lower().str.contains("mixed").sum()
```

```python
#event contains the string "pairs"
summer.Event.str.lower().str.contains("pairs").sum()
```

```python
summer.Sport.loc[summer.Sport == "Equestrian"]
```

```python
#Sport contains the string "equestrian"
summer.Sport.str.lower().str.contains("equestrian").sum()
```

```python
#All sailing event before 1988 were a mixed event
((summer.Sport.str.contains("Sailing")) & (summer.Year < 1988)).sum()
```

```python
summer["Event_Gender"] = summer.Gender
```

```python
mask1 = summer.Event.str.lower().str.contains("mixed")
mask2 = summer.Event.str.lower().str.contains("pairs")
mask3 = summer.Sport.str.lower().str.contains("equestrian")
mask4 = ((summer.Sport.str.lower().str.contains("sailing")) & (summer.Year < 1988))
```

```python
#Assign all the mixed Event with 'X'
summer.loc[mask1 | mask2 | mask3| mask4, "Event_Gender"] = "X"
```

# Counting Medals in Each Event

```python
summer
```

```python
summer.groupby(["Year", "Sport", "Discipline", "Event", "Event_Gender"]).Medal.count().unstack()
```

```python
summer["Event_Medal_Count"] = summer.groupby(["Year", "Sport", "Discipline", "Event", "Event_Gender"]).Medal.transform("count")
```

```python
summer
```

```python
summer.Event_Medal_Count.value_counts().sort_index()
```

# After couting Medal for each Event, assign if the event is a single or team event

```python
summer["Team"] = pd.Series(np.where(summer.Event_Medal_Count > 5, "Yes", "No"))
```

```python
summer
```

```python
summer.Team.value_counts()
```

# Removing Duplicate from Team Events

```python
summer.reset_index(inplace=True)
```

```python
summer
```

```python
singles = summer.loc[summer.Team == "No"].copy()
singles.shape
```

```python
team = summer.loc[summer.Team == "Yes"].copy()
team.shape
```

```python
team
```

```python
team.drop_duplicates(subset = ["Year", "Sport", "Discipline", "Country", "Event", "Event_Gender", "Medal"], inplace = True)
```

```python
team.shape
```

```python
summer_new = pd.concat([singles, team])
```

```python
summer_new.set_index("index", inplace= True)
```

```python
summer_new
```
