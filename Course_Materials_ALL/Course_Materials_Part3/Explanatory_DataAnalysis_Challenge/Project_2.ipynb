# Most Successful Countries of all times

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
olympic
```

```python
top_10 = olympic.Country.value_counts().head(10)
```

```python
olympic_10 = olympic.loc[olympic.Country.isin(top_10.index)]
olympic_10
```

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=1.5, palette= "dark")
sns.countplot(data = olympic_10, x = "Country", hue = "Edition", order = top_10.index)
plt.title("Top 10 Countries by Medals", fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=1.5, palette= "dark")
sns.countplot(data = olympic_10, x = "Country", hue = "Medal", order = top_10.index,
              hue_order = ["Gold", "Silver", "Bronze"], palette = ["gold", "silver", "brown"])
plt.title("Top 10 Countries by Medals", fontsize = 20)
plt.show()
```

# Do GDP, Population and Politics matter?

```python
olympic
```

```python
pd.crosstab(index=olympic.Country, columns=olympic.Medal,margins=True, margins_name="Total")
```

```python
olympic.groupby(["Country", "Medal"]).Medal.count().unstack(fill_value = 0)
```

```python
medals_per_country = olympic.groupby(["Country", "Medal"]).Medal.count().unstack(fill_value = 0)
```

```python
medals_per_country["Total"] = medals_per_country.sum(axis = 1)
```

```python
medals_per_country
```

```python
medals_per_country = medals_per_country[["Total", "Gold", "Silver","Bronze"]].sort_values(["Total", "Gold", "Silver", "Bronze"],
                                                                                         ascending= False)
```

```python
medals_per_country
```

```python
dictionary
```

```python
medals_per_country.merge(dictionary, how = "left", left_index= True, right_on="Country"
                        ).drop(columns = ["Code"]).set_index("Country")
```

```python
medals_per_country = medals_per_country.merge(dictionary, how = "left", left_index= True,
                                             right_on= "Country").drop(columns= ["Code"]).set_index("Country")
```
