# Merging and Concatenating

```python
pd.concat([summer, winter], keys = ["Summer", "Winter"], names = ["Edition"])
```

```python
pd.concat([summer, winter], keys = ["Summer", "Winter"], names = ["Edition"]).reset_index().drop(columns="level_1")
```

```python
olympic = pd.concat([summer, winter], keys = ["Summer", "Winter"], 
                    names = ["Edition"]).reset_index().drop(columns="level_1")
```

```python
olympic
```

```python
dictionary
```

```python
olympic.merge(dictionary, how = "left", left_on= "Country", right_on= "Code").drop(columns= ["Code", "GDP per Capita"])
```

```python
olympic = olympic.merge(dictionary, "left", left_on= "Country", 
                        right_on= "Code").drop(columns= ["Code", "GDP per Capita"]) 
```

```python
olympic
```

```python
olympic.info()
```

```python
olympic.reset_index(drop= True, inplace= True)
```
